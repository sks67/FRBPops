```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import pairwise_distances_argmin
from sklearn.datasets.samples_generator import make_blobs
```

```python
np.random.seed(0)
centers = [[0, 0], [-1, -1], [-2, -2]]
n_clusters = len(centers)
X, labels_true = make_blobs(n_samples=50, centers=centers, cluster_std=1.5)
plt.plot(X[:,0],X[:,1],'.')
plt.show()
```

```python
k_means = KMeans(init='k-means++', n_clusters=3, n_init=10)
k_means.fit(X)
k_means_labels = k_means.labels_
k_means_cluster_centers = k_means.cluster_centers_
k_means_labels_unique = np.unique(k_means_labels)
print k_means_cluster_centers
```

```python
# Plot result

fig = plt.figure(figsize=(6, 5))
colors = ['#4EACC5', '#FF9C34', '#4E9A06']

# We want to have the same colors for the same cluster from the
# MiniBatchKMeans and the KMeans algorithm. Let's pair the cluster centers per
# closest one.
ax = fig.gca()
# KMeans
for k, col in zip(range(n_clusters), colors):
    my_members = k_means_labels == k
    cluster_center = k_means_cluster_centers[k]
    ax.plot(X[my_members, 0], X[my_members, 1], 'w',
            markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=6)
ax.set_title('KMeans')
#ax.set_xticks(())
#ax.set_yticks(())
#plt.text(-3.5, 1.8,  'train time: %.2fs\ninertia: %f' % (
 #   t_batch, k_means.inertia_))
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
# Generate sample data
centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(n_samples=750, centers=centers, cluster_std=0.4,
                            random_state=0)
plt.figure()
plt.scatter(X[:,0],X[:,1])
print np.std(X,axis=0,ddof=1)

X = StandardScaler().fit_transform(X)   #Seems to transform the dataset nearly to Gaussian distribution of N(0,1)
print np.std(X,axis=0,ddof=1)
plt.figure()
plt.scatter(X[:,0],X[:,1])
print X.shape
```

```python
db = DBSCAN(eps=0.3, min_samples=10).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

print labels
```
